```python
import numpy as np
from pyspark.mllib.clustering import KMeans
import matplotlib.pyplot as plt

random_points = sc.textFile("/home/anant/projects/spark-examples/data/random_points.csv")
random_points.cache()
random_points.takeSample(True, 5)
```

```python
data = random_points.map(lambda row: row.split(",")).map(lambda row: np.array([float(row[0]), float(row[1])]))
data.take(5)
```

```python
centers = 6
model = KMeans.train(data, centers)
```

```python
for i , center in enumerate(model.centers):
    print i, center
```

```python
x = data.map(lambda row: row[0]).collect()
y = data.map(lambda row: row[1]).collect()
plt.scatter(x, y)
colors = ["red", "yellow", "green", "grey", "pink", "black"]
plt.scatter([i[0] for i in model.centers], [i[1] for i in model.centers], s=200, c=colors[:centers], alpha=0.8)
plt.show()
```

```python
predict_points = data.takeSample(True, 5)
p_x = []
p_y = []
groups = []
for point in predict_points:
    p_x.append(point[0])
    p_y.append(point[1])
    groups.append(model.predict(point))
    print point, model.predict(point)
```

```python
predict_colors = [colors[group] for group in groups]
plt.scatter(x, y)
plt.scatter(p_x, p_y, s=160,c=predict_colors)
plt.scatter([i[0] for i in model.centers], [i[1] for i in model.centers], s=300, c=colors[:centers], alpha=0.8)
plt.show()
```
